# character count

```python
Character count:
def character_count(filename):
    counts = {}
    with open(filename, 'r') as file:
        for line in file:
            for ch in line:
                if ch != '\n':
                    counts[ch] = counts.get(ch, 0) + 1
    for char, count in counts.items():
        print(f"{char}: {count}")
character_count("input.txt")
```

# Word count

```python
def word_count(filename):
    counts = {}
    with open(filename, 'r') as file:
        for line in file:
            words = line.split()
            for word in words:
                counts[word] = counts.get(word, 0) + 1
    for word, count in counts.items():
        print(f"{word}: {count}")
word_count("input.txt")
```
